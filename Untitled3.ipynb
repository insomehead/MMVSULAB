```python
# 1 
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint

def dydt(x,t):
  k=0.42
  return k*x

t=np.linspace(0,10,100)
solve = odeint(dydt, 1, t=t)
solve = solve.flatten()
%matplotlib inline
plt.plot(t,solve,label="x'=kx")
plt.grid(True)
plt.legend()
plt.show()

# 2
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint

def k(x):
  a=1
  b=1
  return a*x-b

def dydt(x,t):
  return k(x)*x

t=np.linspace(0,10,100)
solve = odeint(dydt, 1, t=t)
solve = solve.flatten()
%matplotlib inline
plt.plot(t,solve,label="x'= k(x)x")
plt.grid(True)
plt.legend()
plt.show()



# 3
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint

def k(x):
  a=1
  b=1
  return a*x-b

def dydt(x,t):
  c=1
  return k(x)*x-c

t=np.linspace(0,10,100)
solve = odeint(dydt, 1, t=t)
solve = solve.flatten()
%matplotlib inline
plt.plot(t,solve,label=r"x'= k(x)x-c")
plt.grid(True)
plt.legend()
plt.show()
```
