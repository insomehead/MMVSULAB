Модель Мальтиуса роста населения Земли
***x' = -kx***

```python
def dxdt(x,t):
    k=-0.0372
    return -k*x
t=years
solve=odeint(dxdt,y0=1,t=t)
solve=solve.flatten()
%matplotlib inline
plt.plot(years, people, c = 'g')
plt.plot(t,solve,label=r"$x'=-kx$")
plt.grid(True)
plt.legend()
plt.show()
```

Мягкая модель Мальтуса с зависящим от населения коэффициентом размножения
**x'=k(x)*x**

```python
def k(x):
    a = 190000000
    b = 300
    return a - b*x

def dydt(x,t):
    return k(x)*x

solve = odeint(dydt, y0=317, t=years)
solve = solve.flatten()

%matplotlib inline
plt.plot(t, solve, label = r"$y' = ay-by^2$")
plt.plot(years, people, c = 'g')
plt.grid(True)
plt.legend()
plt.show()
```
